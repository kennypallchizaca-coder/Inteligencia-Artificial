Una lista es un conjunto de elementos de cualquier tipo. Una lista es dinámica (mutable). Una tupla es un conjunto ordenado e inmutable de elementos.

```python
numeros=[10,20,30,40,50]
print("Lista completa:", numeros)
print("Primer elemento:", numeros[0])
print("Último elemento:", numeros[-1])
print("Tamaño de la lista:", len(numeros))
```

```python
numeros.append(60)
print("Lista después de agregar un elemento:", numeros)
```

Un diccionario es un tipo de estructura de datos que permite guardar un conjunto no ordenado de pares clave-valor, siendo las claves únicas dentro de un mismo diccionario.

```python
estudiante ={
    "nombre": "Alexis",
    "edad": 20,
    "Carrera": "Computación"
}

print(estudiante)
print("Nombre del estudiante:", estudiante["nombre"])
print("Edad del estudiante:", estudiante["edad"])
print("Carrera del estudiante:", estudiante["Carrera"])
```

```python
# Aplanar un diccionario jerárquico (nested dict/map)
from typing import Any

nested_data = {
    "a": {
        "b": "1",
        "c": {
            "d": "2",
            "e": "3",
        },
    },
    "f": 4,
    "g": {
        "h": 5, 
        "i": 6,
        "j": 7, 
        "k": 8,
    },
}
separator: str="/"
result: dict[str, Any] = {}

def flatten(current_value: Any, current_key: str) -> None:

    if isinstance(current_value, dict):
        for key, value in current_value.items():
            key_str = str(key)
            value_str = str(value)
            new_key = (
                f"{current_key}{separator}{key_str}"
                if current_key
                else key_str
            )
            flatten(value, new_key)
        return
    result[current_key] = current_value

flatten(nested_data, "")

print("\ndict aplanado completo:\n")
print(result)
```

```python
# Dividir una cadena en palabras
s = "Python es el lenguaje de preferencia de las ingenierías!"
s = s.split()
print("Split:", s)

# Obtener nombre de dominio
def obtenerDominio(s='rhurtadoo@ups.edu.ec'):
    return s[s.find('@')+1:]
print("Dominio:", obtenerDominio())

# Operaciones con listas superiores
seq = ['datos','internet','dado','protocolo', 'driver']
listaFiltrada = list(filter(lambda k: 'd' in k ,seq))
print("Filtro empieza con d:", listaFiltrada)

listaMayusculas = list(map(lambda x:x.upper(), seq))
print("Map a mayúsculas:", listaMayusculas)
```

```python
# La función zip (lista1, lista2) devuelve una lista de tuplas
pisos = ['Estacionamiento', 'Tiendas', 'Patio de comidas', 'Oficinas']
pisos_numeros = range(-1,3)

diccionario = dict(zip(pisos_numeros,pisos)) 
print("Diccionario generado con zip:", diccionario)
print("Piso -1 es:", diccionario[-1])
```

```python
# Estructura condicional if
edad = 20

if edad >= 18:
    print("Eres mayor de edad.")
else:
    print("Eres menor de edad.")
```

```python
numeros_b = [10, 20, 30]
# Estructura iterativa for
for numero in numeros_b:
    print("Numero:", numero)
```

```python
# Estructura iterativa while
contador = 1 

while contador <= 5:
    print("Contador:", contador)
    contador += 1
```
