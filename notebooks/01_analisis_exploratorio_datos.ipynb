```python
variables_outliers = [
    "edad",
    "presion_arterial_reposo",
    "colesterol",
    "frecuencia_cardiaca_max",
    "oldpeak"
]

for columna in variables_outliers:
    plt.figure(figsize=(7, 4))
    sns.boxplot(x=df[columna])
    plt.title(f"Boxplot de {columna}")
    plt.xlabel(columna)
    plt.show()
```

```python
def detectar_outliers_iqr(dataframe, columna):
    q1 = dataframe[columna].quantile(0.25)
    q3 = dataframe[columna].quantile(0.75)
    iqr = q3 - q1
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr
    
    outliers = dataframe[
        (dataframe[columna] < limite_inferior) |
        (dataframe[columna] > limite_superior)
    ]
    
    return len(outliers), limite_inferior, limite_superior
```

```python
for columna in variables_outliers:
    cantidad, li, ls = detectar_outliers_iqr(df, columna)
    print(f"{columna}: {cantidad} outliers | límite inferior = {li:.2f} | límite superior = {ls:.2f}")
```
