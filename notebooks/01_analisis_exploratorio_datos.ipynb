Las variables categóricas requieren un análisis de frecuencia específico para entender cómo se distribuyen sus categorías y cómo se relacionan con la variable objetivo.

Se analizan las siguientes variables categóricas del dataset:
- **SEXO** (sex): male / female
- **CLASE** (pclass): 1ra, 2da, 3ra clase del boleto
- **PUERTO_EMBARQUE** (embarked): C = Cherbourg, Q = Queenstown, S = Southampton
- **QUIEN** (who): man, woman, child
- **CUBIERTA** (deck): A-G (descartada por 77% de valores faltantes)

Para cada variable se calcula la **tabla de frecuencias** y la **tasa de supervivencia por categoría**, para identificar cuáles grupos tienen mayor o menor probabilidad de haber sobrevivido.

```python
def descripcionCantidadCategorias(dataframe, categorical_features):
    cantidadVariablesCategoricas = len(categorical_features)
    for variable in categorical_features:
        cantidadCategorias = dataframe[variable].nunique()
        print('Cantidad de categorías en la variable', variable, ':', cantidadCategorias)
    return cantidadVariablesCategoricas
```

```python
categorical_features = ['SEXO', 'CLASE', 'PUERTO_EMBARQUE', 'QUIEN', 'CUBIERTA']

cantidadVariablesCategoricas = descripcionCantidadCategorias(dataframe, categorical_features)
print('Cantidad de variables categóricas:', cantidadVariablesCategoricas)
```

El análisis de correlación permite identificar qué variables numéricas tienen una relación lineal con la variable objetivo `SOBREVIVIO` y entre sí.

Se utilizan:
- **Coeficiente de correlación de Pearson (r):** mide la dirección e intensidad de la relación lineal
- **Mapa de calor (heatmap):** visualización de la matriz de correlación completa

Interpretación del coeficiente r:
- |r| > 0.5 → correlación fuerte
- 0.3 < |r| < 0.5 → correlación moderada
- |r| < 0.3 → correlación débil

Se crean variables numéricas derivadas (`SEXO_NUM`, `TAMANO_FAMILIA`, `VIAJA_SOLO`) para incluirlas en el análisis de correlación con la variable objetivo.

```python
# Crear variables numéricas derivadas para el análisis de correlación
dataframe['SEXO_NUM'] = (dataframe['SEXO'] == 'male').astype(int)
dataframe['TAMANO_FAMILIA'] = dataframe['HERMANOS_CONYUGES'] + dataframe['PADRES_HIJOS'] + 1
dataframe['VIAJA_SOLO'] = (dataframe['TAMANO_FAMILIA'] == 1).astype(int)

x = np.array([5, 5, 1])
y = np.array([1, 2, 5])
covariance = np.cov(x, y, bias=True)[0][1]
print('Covarianza:', covariance)
dict_ejemplo = {'X': x, 'Y': y}
df_ejemplo = pd.DataFrame(dict_ejemplo)
std = df_ejemplo['Y'].std(ddof=0)
print('Desviación estándar:', std)
df_ejemplo.corr()
```

```python
#Análisis entre todas las variables numéricas
variablesCorrelacion = ['SOBREVIVIO', 'CLASE', 'EDAD', 'HERMANOS_CONYUGES', 
                        'PADRES_HIJOS', 'TARIFA', 'SEXO_NUM', 'TAMANO_FAMILIA', 'VIAJA_SOLO']
cr = dataframe[variablesCorrelacion].corr()
cr.head()
```

```python
#Mapa de calor
sns.set(style='white')
mask = np.zeros_like(cr, dtype=np.bool_)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(cr, mask=mask, cmap=cmap, vmax=.3, center=0,
            annot=True, fmt='.2f',
            square=True, linewidths=.5, cbar_kws={'shrink': .5})
plt.title('Mapa de calor de correlaciones')
plt.show()
```

```python
# Identificar las 3 variables más correlacionadas con SOBREVIVIO
correlacion_objetivo = cr['SOBREVIVIO'].drop('SOBREVIVIO').abs().sort_values(ascending=False)
print('Top 3 variables más correlacionadas con SOBREVIVIO:')
print('=' * 50)
for i, (var, val) in enumerate(correlacion_objetivo.head(3).items(), 1):
    raw = cr['SOBREVIVIO'][var]
    direccion = 'Positiva' if raw > 0 else 'Negativa'
    print(f'{i}. {var:<20} r = {raw:+.3f} ({direccion})')
```

**Conclusiones del análisis de correlación:**

- **SEXO_NUM (r ≈ -0.54):** Es la variable con mayor correlación (negativa) con SOBREVIVIO. Ser masculino disminuye significativamente la probabilidad de supervivencia.
- **CLASE (r ≈ -0.34):** Correlación negativa moderada-alta. A mayor número de clase (mayor pobreza), menor probabilidad de sobrevivir.
- **TARIFA (r ≈ +0.26):** Correlación positiva moderada. A mayor tarifa pagada, mayor probabilidad de supervivencia (proxy del nivel socioeconómico).

Variables con correlación lineal débil:
- **EDAD (r ≈ -0.06):** La edad por sí sola no determinó el resultado de supervivencia.
- **TAMANO_FAMILIA (r ≈ -0.02):** Correlación casi nula con el método lineal.

Los valores faltantes pueden introducir sesgos en el análisis y afectar el rendimiento de los modelos predictivos. Se identifican y tratan con estrategias específicas según el porcentaje de nulos y la naturaleza de cada variable:

| Variable | Nulos | Estrategia |
|---|---|---|
| EDAD (age) | 177 (19.8%) | Imputación con la **mediana** (robusta ante outliers) |

```python
# Resumen de datos faltantes
print('Valores faltantes antes del tratamiento:')
print(dataframe.isnull().sum()[dataframe.isnull().sum() > 0])
```

```python
# Estrategia de manejo:
# 1. EDAD: 177 nulos (19.8%) => imputar con la mediana (robusta ante outliers)
mediana_edad = dataframe['EDAD'].median()
dataframe['EDAD'] = dataframe['EDAD'].fillna(mediana_edad)
print(f'EDAD: 177 nulos imputados con mediana = {mediana_edad:.1f} años')

# 2. CUBIERTA: 688 nulos (77.2%) => eliminar columna (demasiados nulos)
dataframe = dataframe.drop(columns=['CUBIERTA'])
print('CUBIERTA: columna eliminada (77.2% de nulos)')

# 3. PUERTO_EMBARQUE: 2 nulos => imputar con la moda
moda_puerto = dataframe['PUERTO_EMBARQUE'].mode()[0]
dataframe['PUERTO_EMBARQUE'] = dataframe['PUERTO_EMBARQUE'].fillna(moda_puerto)
print(f'PUERTO_EMBARQUE: 2 nulos imputados con moda = {moda_puerto}')

print(f'\nValores faltantes restantes: {dataframe.isnull().sum().sum()}')
print(f'Dataset limpio: {dataframe.shape[0]} filas x {dataframe.shape[1]} columnas')
```

```python
# Verificar variables redundantes o no relevantes para el análisis
# VIVO es equivalente a SOBREVIVIO (texto vs numérico)
# CLASE_CAT es equivalente a CLASE (categorica vs numérica)
print('Verificación de equivalencias:')
print(pd.crosstab(dataframe['VIVO'], dataframe['SOBREVIVIO']))
```

    <h3 style="margin: 0; text-transform: uppercase;">&gt; Análisis y eliminación de outliers_ &#9608;</h3>

Los **valores atípicos (outliers)** son observaciones que se alejan significativamente del resto de los datos. Su presencia puede distorsionar el análisis estadístico y afectar el rendimiento de ciertos algoritmos de machine learning.

Se utiliza el **método de Tukey (IQR)** para la detección:
- **Límite inferior:** Q1 - 1.5 × IQR
- **Límite superior:** Q3 + 1.5 × IQR

Se analizan las variables: `EDAD`, `TARIFA`, `HERMANOS_CONYUGES`, `PADRES_HIJOS`

```python
# Detección de outliers con el método IQR (Tukey)
def detectarOutliers(dataframe, variable):
    Q1 = dataframe[variable].quantile(0.25)
    Q3 = dataframe[variable].quantile(0.75)
    IQR = Q3 - Q1
    limiteInferior = Q1 - 1.5 * IQR
    limiteSuperior = Q3 + 1.5 * IQR
    outliers = dataframe[(dataframe[variable] < limiteInferior) | 
                         (dataframe[variable] > limiteSuperior)]
    print(f'Variable: {variable}')
    print(f'  Q1={Q1:.2f}, Q3={Q3:.2f}, IQR={IQR:.2f}')
    print(f'  Límite inferior: {limiteInferior:.2f}')
    print(f'  Límite superior: {limiteSuperior:.2f}')
    print(f'  Outliers detectados: {len(outliers)} ({len(outliers)/len(dataframe)*100:.1f}%)')
    print(f'  Valor máximo: {dataframe[variable].max():.2f}')
    print()
    return limiteInferior, limiteSuperior
```

```python
variablesOutliers = ['EDAD', 'TARIFA', 'HERMANOS_CONYUGES', 'PADRES_HIJOS']
for v in variablesOutliers:
    detectarOutliers(dataframe, v)
```

```python
# Visualización de outliers con diagramas de caja
fig, axes = plt.subplots(1, 4, figsize=(16, 5))
for i, var in enumerate(variablesOutliers):
    dataframe.boxplot(column=var, ax=axes[i])
    axes[i].set_title(var)
plt.suptitle('Análisis de outliers (método IQR)')
plt.tight_layout()
plt.show()
```

**Conclusiones del análisis de outliers:**

Se detectaron valores atípicos principalmente en las variables **TARIFA** (~13% de outliers, con valores hasta 512£) y moderados en **EDAD** (~2%, con individuos de hasta 80 años).

**Decisión: NO eliminar los outliers. Justificación:**
1. Los valores extremos de TARIFA corresponden a pasajeros de 1ra clase con suites de lujo — son datos históricos fidedignos, no errores de medición.
2. Los outliers de TARIFA están directamente correlacionados con la supervivencia. Eliminarlos distorsionaría el modelo predictivo.
3. El dominio es histórico: cada observación representa una persona real documentada. Censurar estos datos viola la integridad histórica del análisis.
4. Los algoritmos basados en árboles de decisión (Random Forest, XGBoost) son intrínsecamente robustos ante outliers.

La segmentación divide el dataset en grupos con características similares para un análisis más granular. Permite identificar patrones que no son visibles en el análisis global del dataset.

Se realizan los siguientes análisis de segmentación:
1. **Tabla cruzada SEXO × CLASE:** tasa de supervivencia para cada combinación
2. **Segmentación por tamaño familiar:** Solo / Familia Pequeña (2-4) / Familia Grande (5+)

El resultado permite identificar los segmentos con mayor y menor probabilidad de supervivencia.

```python
# Segmentación por sexo y clase: análisis cruzado de supervivencia
pivotSupervivencia = dataframe.pivot_table(
    values='SOBREVIVIO', index='SEXO', columns='CLASE', aggfunc='mean'
)
print('Tasa de supervivencia (%) por SEXO y CLASE:')
print((pivotSupervivencia * 100).round(1))

sns.heatmap(pivotSupervivencia, annot=True, fmt='.1%',
            cmap='RdYlGn', vmin=0, vmax=1,
            linewidths=0.5)
plt.title('Tasa de Supervivencia por SEXO y CLASE')
plt.show()
```

```python
# Segmentación por tamaño de familia
def clasificarFamilia(tamano):
    if tamano == 1:
        return 'Solo'
    elif tamano <= 4:
        return 'Familia Pequeña (2-4)'
    else:
        return 'Familia Grande (5+)'

dataframe['GRUPO_FAMILIA'] = dataframe['TAMANO_FAMILIA'].apply(clasificarFamilia)

print('Tasa de supervivencia por GRUPO_FAMILIA:')
print((dataframe.groupby('GRUPO_FAMILIA')['SOBREVIVIO'].mean() * 100).round(1))

sv_familia = dataframe.groupby('GRUPO_FAMILIA')['SOBREVIVIO'].mean() * 100
sv_familia = sv_familia.reindex(['Solo', 'Familia Pequeña (2-4)', 'Familia Grande (5+)'])
sv_familia.plot(kind='bar', color=['steelblue', 'green', 'salmon'], rot=20)
plt.title('Tasa de Supervivencia por Tamaño de Familia')
plt.ylabel('Tasa de supervivencia (%)')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```
